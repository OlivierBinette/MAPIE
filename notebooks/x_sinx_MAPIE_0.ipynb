# I. Estimating the aleatoric uncertainty of homoscedastic noisy data

In this section, we focus on homoscedastic data.

## Generating the dataset

We first generate a 1-dimensional dataset.

```python
noise = 0.5
X_train, y_train, X_test, y_test, y_mesh = get_1d_data_with_constant_noise(
    func=lambda x: x*np.sin(x),
    minimum=-5,
    maximum=5,
    n_samples=100,
    noise=noise
)
```

This is nothing but the function $x \mapsto x\sin(x)$ with additional noise.

```python
fig = go.Figure()
fig.add_trace(
    go.Scatter(
        x=X_train.ravel(),
        y=y_train,
        name="Training set",
        mode="markers",
        marker=dict(color="blue")
    )
)
fig.add_trace(
    go.Scatter(
        x=X_test.ravel(),
        y=y_mesh,
        name="Truth",
        mode="lines",
        line=dict(color="red")
    )
)
fig.update_layout(
    xaxis_title="x",
    yaxis_title="y",
    hovermode="x",
    font=dict(family="Computer Modern", size=20, color="#7f7f7f")
)
```

## Defining the base regressors

We are going to compare three base models, namely :
* polynomial regression
* random forest
* neural network

```python
models = {
    "polynomial": Pipeline(
        [
            ("poly", PolynomialFeatures(degree=10)),
            ("linear", LinearRegression(fit_intercept=False))
        ]
    ),
    "random_forest": RandomForestRegressor(),
    "neural_network": MLPRegressor(max_iter=2000)
}
```

## Estimating prediction intervals

```python
y_pred, y_pis = {}, {}
for name, model in models.items():
    mapie = MapieRegressor(model, method="plus", cv=KFold(shuffle=True))
    mapie.fit(X_train, y_train)
    y_pred[name], y_pis[name] = mapie.predict(X_test, alpha=0.05)
```

```python
fig = make_subplots(rows=1, cols=3)
for col, name in enumerate(models):
    fig.add_trace(
        go.Scatter(
            x=X_train.ravel(),
            y=y_train,
            name="Training set",
            mode="markers",
            marker=dict(color="blue"),
            showlegend=False
        ),
        row=1,
        col=col + 1
    )
    fig.add_trace(
        go.Scatter(
            x=X_test.ravel(),
            y=y_mesh,
            name="Truth",
            line=dict(color="red"),
            showlegend=False
        ),
        row=1,
        col=col + 1
    )
    fig.add_trace(
        go.Scatter(
            x=X_test.ravel(),
            y=y_mesh - 1.96*noise,
            line=dict(color="red", dash="dot"),
            showlegend=False
        ),
        row=1,
        col=col + 1
    )
    fig.add_trace(
        go.Scatter(
            x=X_test.ravel(),
            y=y_mesh + 1.96*noise,
            line=dict(color="red", dash="dot"),
            showlegend=False
        ),
        row=1,
        col=col + 1
    )
    fig.add_trace(
        go.Scatter(
            x=X_test.ravel(),
            y=y_pred[name],
            name=name,
            mode="lines",
            marker=dict(color="green"),
            showlegend=True
        ),
        row=1,
        col=col + 1
    )
    fig.add_trace(
        go.Scatter(
            name="lower bound",
            x=X_test.ravel(),
            y=y_pis[name][:, 0],
            mode="lines",
            line=dict(color="rgba(0, 255, 0, 0.3)", dash='solid')
        )
    )
    fig.add_trace(
        go.Scatter(
            name="upper bound",
            x=X_test.ravel(),
            y=y_pis[name][:, 1],
            fill="tonexty",
            fillcolor="rgba(0, 0, 255, 0.1)",
            line=dict(color="rgba(0, 255, 0, 0.3)", dash='solid')
        )
    )
fig.update_layout(
    xaxis_title="x",
    yaxis_title="y",
    hovermode="x",
    font=dict(family="Computer Modern", size=12, color="#7f7f7f")
)
```

# II. Estimating aleatoric and epistemic uncertainties with different methods

## Generating the dataset

```python
mu, sigma, n_samples, noise = 0, 2.5, 300, 0.5
X_train, y_train, X_test, y_test, y_mesh = get_1d_data_with_normal_distrib(x_sinx, mu, sigma, n_samples, noise)
```

```python
plt.figure(figsize=(10, 8))
plt.xlabel("x") ; plt.ylabel("y")
plt.scatter(X_train, y_train, color="C0")
plt.plot(X_test, y_mesh, color="C1")
```

## Estimating the prediction intervals

```python
# Definition of our base model
degree_polyn = 10
polyn_model = Pipeline(
    [
        ("poly", PolynomialFeatures(degree=degree_polyn)),
        ("linear", LinearRegression())
    ]
)
```

```python
# Estimating prediction intervals
Params = TypedDict("Params", {"method": str, "agg_function":str, "cv": Union[int, Subsample]})
STRATEGIES = {
    "Jackknife+": Params(method="plus", cv=-1),
    "Jackknife-minmax": Params(method="minmax", cv=-1),
    "CV+": Params(method="plus", cv=10),
    "CV-minmax": Params(method="minmax",  cv=10),
    "Jackknife+ after Bootstrap": Params(method="plus", agg_function="mean", cv=Subsample(n_resamplings=50)),
    "Jackknife-minmax after Bootstrap": Params(method="minmax", agg_function="mean", cv=Subsample(n_resamplings=50)),
}
```

```python
y_pred, y_pis, pred_multi = {}, {}, {}
for strategy, params in STRATEGIES.items():
    mapie = MapieRegressor(polyn_model, **params)
    mapie.fit(X_train, y_train)
    y_pred[strategy], y_pis[strategy] = mapie.predict(X_test, alpha=0.05)
```

```python
n_figs = len(STRATEGIES)
fig, axs = plt.subplots(n_figs//2, 2, figsize=(12, 14))
coords = []
for xaxis in range(n_figs//2):
    coords.append(axs[xaxis, 0])
    coords.append(axs[xaxis, 1])
    
for strategy, coord in zip(STRATEGIES, coords):
    plot_1d_data(
        X_train.ravel(),
        y_train.ravel(),
        X_test.ravel(),
        y_mesh.ravel(),
        1.96*noise,
        y_pred[strategy].ravel(),
        y_pis[strategy][:, 0, 0].ravel(),
        y_pis[strategy][:, 1, 0].ravel(),
        # pred_multi[strategy],
        ax=coord,
        title=strategy,
        set_limits=True
    )
```

```python
fig, ax = plt.subplots(1, 1, figsize=(7, 5))
ax.set_xlim([-8, 8])
ax.set_ylim([1.5, 4])
for strategy in STRATEGIES:
    ax.plot(X_test, y_pis[strategy][:, 1, 0] - y_pis[strategy][:, 0, 0])
ax.axhline(1.96*2*noise, ls="--", color="k")
ax.set_xlabel("x")
ax.set_ylabel("Prediction Interval Width")
ax.legend(list(STRATEGIES.keys()) + ["True width"], fontsize=8)
plt.show()
```

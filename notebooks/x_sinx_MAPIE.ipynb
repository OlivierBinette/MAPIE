# MAPIE Demonstration

In this tutorial, we are going to investigate some of MAPIE properties.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import KFold

from mapie.subsample import Subsample 
from mapie.regression import MapieRegressor
from mapie.metrics import regression_coverage_score

from utils import (
    x_sinx,
    get_1d_data_with_constant_noise,
    get_1d_data_with_normal_distrib,
    plot_1d_data
)

%load_ext autoreload
%autoreload 2
```

# I. Estimating the aleatoric uncertainty of homoscedastic noisy data

In this section, we focus on homoscedastic data.

## Generating the dataset

We first generate a 1-dimensional dataset.

```python
noise = 0.5
X_train, y_train, X_test, y_test, y_mesh = get_1d_data_with_constant_noise(
    func=lambda x: x*np.sin(x),
    minimum=-5,
    maximum=5,
    n_samples=300,
    noise=noise
)
```

This is nothing but the function $x \mapsto x\sin(x)$ with additional noise.

```python
plt.figure(figsize=(10, 8))
plt.xlabel("x") ; plt.ylabel("y")
plt.scatter(X_train, y_train, color="C0")
plt.plot(X_test, y_mesh, color="C1")
```

## Defining the base regressors

We are going to compare three base models, namely :
* polynomial regression
* random forest
* neural network

```python
polyn_model = Pipeline(
    [
        ("poly", PolynomialFeatures(degree=10)),
        ("linear", LinearRegression(fit_intercept=False))
    ]
)
rf_model = RandomForestRegressor()
mlp_model = MLPRegressor(max_iter=2000)
```

Let us rearrange them into a list.

```python
models = [polyn_model, rf_model, mlp_model]
model_names = ["Polynomial function", "Random Forest", "Neural Network"]
```

## Estimating prediction intervals

At this point, it is as simple to use MAPIE as it is to use any scikit-learn estimator.

```python
y_pred, y_pis = {}, {}
for name, model in zip(model_names, models):
    mapie = MapieRegressor(model, method="plus", cv=KFold(shuffle=True))
    mapie.fit(X_train, y_train)
    y_pred[name], y_pis[name] = mapie.predict(X_test, alpha=0.05)
```

We can plot the prediction intervals of all estimators.

```python
fig, axs = plt.subplots(1, len(models), figsize=(20, 6))
for name, ax in zip(model_names, axs):
    plot_1d_data(
        X_train.ravel(),
        y_train.ravel(),
        X_test.ravel(),
        y_mesh.ravel(),
        1.96*noise,
        y_pred[name].ravel(),
        y_pis[name][:, 0, 0].ravel(),
        y_pis[name][:, 1, 0].ravel(),
        ax=ax,
        title=name+" with CV+"
    )
```

## Checking the calibration of the CV+ method

In this section we investigate the relationship between expected and measured coverages as a function of alpha.

We thus define many value of alpha, between 0 and 1.

```python
alpha = np.arange(0.05, 0.95, 0.02)
```

MAPIE allows the user to use all alphas at once at prediction time.

```python
mapie = MapieRegressor(polyn_model, method="plus", cv=KFold(shuffle=True))
mapie.fit(X_train, y_train)
y_pred, y_pis = mapie.predict(X_test, alpha=alpha)
```

As soon as we have our results, we can compute metrics on prediction intervals, namely :
* coverage : the proportion of points truly residing within the prediction interval
* width : the average width of the prediction interval

```python
coverages = [
    regression_coverage_score(y_test, y_pis[:, 0, i], y_pis[:, 1, i])
    for i, _ in enumerate(alpha)
]
widths = [
    np.mean(y_pis[:, 1, i] - y_pis[:, 0, i])
    for i, _ in enumerate(alpha)
]
```

We see that MAPIE is approximately well calibrated, i.e. that the empirial coverage is aligned with expected coverage. Because we are using few folds of cross-validation, we slightly overcoverage testing data.

```python
fig, ax = plt.subplots(1, 2, figsize=(11, 5))
ax[0].set_xlabel("1 - alpha")
ax[0].set_ylabel("Effective coverage on test set")
ax[0].plot([0, 1], [0, 1], ls="--", color="k")
ax[0].scatter(1-alpha, coverages)
ax[1].set_xlabel("1 - alpha")
ax[1].set_ylabel("Average of prediction interval widths")
ax[1].scatter(1-alpha, widths)
```

# II. Estimating aleatoric and epistemic uncertainties with different methods

In this section, we investigate heteroscedastic data

## Generating the dataset

We first generate a 1-dimensional dataset with long-tailed distribution.

```python
noise = 0.5
X_train, y_train, X_test, y_test, y_mesh = get_1d_data_with_normal_distrib(
    func=lambda x: x*np.sin(x),
    mu=0,
    sigma=2.5,
    n_samples=300,
    noise=noise
)
```

This is nothing but the function $x \mapsto x\sin(x)$ with additional noise.

```python
plt.figure(figsize=(10, 8))
plt.xlabel("x") ; plt.ylabel("y")
plt.scatter(X_train, y_train, color="C0")
plt.plot(X_test, y_mesh, color="C1")
```

## Estimating the prediction intervals

We are going to compare different prediction interval methods.

```python
STRATEGIES = {
    "Jackknife+": dict(method="plus", cv=-1),
    "Jackknife-minmax": dict(method="minmax", cv=-1),
    "Jackknife+ after Bootstrap": dict(method="plus", agg_function="mean", cv=Subsample(n_resamplings=50)),
    "Jackknife-minmax after Bootstrap": dict(method="minmax", agg_function="mean", cv=Subsample(n_resamplings=50)),
}
```

We take full advantage of the scikit-learn API.

```python
y_pred, y_pis, pred_multi = {}, {}, {}
for strategy, params in STRATEGIES.items():
    mapie = MapieRegressor(polyn_model, **params)
    mapie.fit(X_train, y_train)
    y_pred[strategy], y_pis[strategy] = mapie.predict(X_test, alpha=0.05)
```

We see that some methods are better than others at capturing epistemic uncertainty.

```python
n_figs = len(STRATEGIES)
fig, axs = plt.subplots(n_figs//2, 2, figsize=(12, 14))
coords = []
for xaxis in range(n_figs//2):
    coords.append(axs[xaxis, 0])
    coords.append(axs[xaxis, 1])
    
for strategy, coord in zip(STRATEGIES, coords):
    plot_1d_data(
        X_train.ravel(),
        y_train.ravel(),
        X_test.ravel(),
        y_mesh.ravel(),
        1.96*noise,
        y_pred[strategy].ravel(),
        y_pis[strategy][:, 0, 0].ravel(),
        y_pis[strategy][:, 1, 0].ravel(),
        # pred_multi[strategy],
        ax=coord,
        title=strategy,
        set_limits=True
    )
```

```python
fig, ax = plt.subplots(1, 1, figsize=(7, 5))
ax.set_xlim([-8, 8])
ax.set_ylim([1.5, 4])
for strategy in STRATEGIES:
    ax.plot(X_test, y_pis[strategy][:, 1, 0] - y_pis[strategy][:, 0, 0])
ax.axhline(1.96*2*noise, ls="--", color="k")
ax.set_xlabel("x")
ax.set_ylabel("Prediction Interval Width")
ax.legend(list(STRATEGIES.keys()) + ["True width"], fontsize=8)
plt.show()
```
